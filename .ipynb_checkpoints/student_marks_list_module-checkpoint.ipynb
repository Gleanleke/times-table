```python
import student_marks_list_module


studentMarks=[]
numOfStudents=int(input("enter the number of student/s in your class "))

x=0
y=1
for x in range(0,numOfStudents):
    def n(y):
        if y==1:
            return "1st" 
        if y==2:
            return "2nd" 
        if y==3:
            return "3rd" 
        if y>3:
            return f"{y}th"
    studentMark=(int(input(f"enter the score of the {n(y)} student ")))
    if studentMark<=20:
        studentMarks.append(studentMark)
    else:
        print()
        print("Student marks must not be greater than 20")
        break
    
    y+=1
print()
if len(studentMarks)>0:
    student_marks_list_module.studentMarksList(studentMarks,numOfStudents)

```
