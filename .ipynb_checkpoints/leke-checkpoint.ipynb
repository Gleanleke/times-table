```python
#problem3
def solveQ(a,b,c):
    x=3
    n=0
    p,q,r,s=a,0,0,c
    i,y,j,e,n=0,0,0,1,1
    z=0
    m=0
    while i<=a*c:
        y=0
        while y<=b:
            j=0
            while j<=b:
                if y+j==b and y*j==(a*c):
                    q=j
                    r=y
                    p,q,r,s=a,j,y,c
                    e,n=1,1
                    while e<=p and e<=q:
                        if p%e==0 and q%e==0:
                            p,q=p//e,q//e
                            z=(e*x*(p*x+q))
                        e+=1
                    
                    while n <=r and n <=s :
                        if r%n==0 and s%n==0:
                            r,s=r//n,s//n
                            m=(n*(r*x+s))
                        n+=1
                    if (r * x + s) == (p * x + q):
                            root1=-q/p
                            root2=-s/r
                            print(f"({p}x + {q}) and ({r}x + {s})=0 ie ({p}x + {q})({r}x + {s})=0 ")
                            print(f"Roots are: x={root1:.4f} and x={root2:.4f}")
                    return
                j+=1
            y+=1
        i+=1
a = int(input("input a: "))
b = int(input("input b: "))
c = int(input("input c: "))

solveQ(a,b,c)
```

```python
def mPaymentOfL(amt,r,n):
    return amt*(r*1200)*(1+(r*1200))*n*(1+(r*1200)*n-1)


amt=int(input("input your ammount "))
interestRate=int(input("whats the interest rate? "))/100
n=int(input("enter the duration in months "))

print(mPaymentOfL(amt,interestRate,n))
```
